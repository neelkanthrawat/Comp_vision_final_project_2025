```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# In domain
data_percentate = [0.25, 0.5, 0.75, 1.0]
localized_lora = [0.7824, 0.7915, 0.7945, 0.7970]
localied_lora_error = [0.0037, 0.0028, 0.0018, 0.0018]
lora = [0.7763, 0.7800, 0.7812, 0.7830]
lora_error = [0.0014, 0.0036, 0.0073, 0.0022]
serial_lora = [0.7569, 0.7643, 0.7744,0.7647]
serial_lora_error = [0.0179, 0.0079, 0.0019, 0.0143]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 5))
ax.errorbar(data_percentate, localized_lora, localied_lora_error, label="Localized Lora", fmt='o', capsize=4, markersize=6,)
ax.errorbar(data_percentate, serial_lora, serial_lora_error, label="Serial Lora", fmt='s', capsize=4, markersize=6,)
ax.errorbar(data_percentate, lora, lora_error, label="Lora", fmt='^', capsize=4, markersize=6,)
ax.set_xticks(data_percentate)
ax.set_xlabel("Data Percentage (%)", fontsize=12)
ax.set_ylabel("Dice Score", fontsize=12)
ax.legend()
ax.tick_params(axis='both', which='major', labelsize=11)
ax.set_ylim(0.7, 0.8)
ax.grid()

fig.tight_layout()
plt.show()
```

```python
# Out of domain
style = ["candy", "mosaic", "rain princess", "udnie"]
lora_mean = [0.5438188638687134, 0.26926658749580384, 0.5064485218524932, 0.6559063384532929]
lora_error = [0.03505749309015839, 0.08787092315557644, 0.03547103853482523, 0.015975295222327164]

serial_lora_mean = [0.574192754983902, 0.2981011798381806, 0.5362250218391419, 0.6529047112464905]
serial_lora_error = [0.04726289187398069, 0.03887903015456093, 0.05121379677405659, 0.024526153688404495]

localized_lora_mean = [0.5477752623558045, 0.2436222208738327, 0.4517842950820922, 0.6610252003669739]
localized_lora_error = [0.03953368045702863, 0.030176790196681954, 0.03845178981166709, 0.015743880766867908]
```

```python
x = np.arange(len(style))
width = 0.25

fig, ax = plt.subplots(1, 1, figsize=(9, 6))
ax.bar(x - width, lora_mean, width,yerr=lora_error, capsize=4, label="Lora")
ax.bar(x, serial_lora_mean, width, yerr=serial_lora_error, capsize=4, label="Serial Lora")
ax.bar(x + width, localized_lora_mean, width, yerr=localied_lora_error, capsize=4, label="Localized Lora")
ax.set_xlabel("Out-of-Domain Style", fontsize=12)
ax.set_ylabel("Dice Score", fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(style, fontsize=11)
ax.set_ylim(0, 0.8)
ax.legend()
fig.tight_layout()
plt.show()
```
